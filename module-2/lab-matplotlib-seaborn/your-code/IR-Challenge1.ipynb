```python
# your code here-1st way (call `subplots` twice using the `index` parameter)
fig, [ax1, ax2] = plt.subplots(1,2) #the 1,2 means 1 row 2 columns, 2,1 would mean 2 rows 1 column
ax1.plot(x,y)
ax1.set_title('Yes')
ax1.set_xlabel('x label')
ax1.set_ylabel('y label') 

ax2.plot(x,z)
ax2.set_title('No')
ax2.set_xlabel('x label')
ax2.set_ylabel('y label')

plt.show()
```

#### Plot both `y=x^2` and `y=exp(x)`in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here
x = np.arange(0,100)
y = x^2
z = np.exp(x)

fig, ax1 = plt.subplots()
ax1.set_xlabel('Doubt')
ax1.set_xscale('linear')
ax1.set_ylabel('Uncertainty')
ax1.plot(x, y, color = 'red')
ax1.tick_params(axis='y')
ax2 = ax1.twinx()
ax2.set_ylabel('Unsure')
ax2.set_yscale("log")
ax2.plot(x,z)
ax2.tick_params(axis='y')
plt.title("Will this Work?")
plt.legend()

plt.show()
```
