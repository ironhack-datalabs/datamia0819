#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here

t = np.arange(0.01, 10.0, 0.01)
data1 = np.exp(t)
data2 = t**2
fig, ax1 = plt.subplots()
ax1.set_xlabel('test')
ax1.set_ylabel('exp', color='red')
ax1.plot(t, data1, color='red')
ax1.tick_params(axis='y')
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
ax2.set_ylabel('log', color='b')
ax2.plot(t, data2, color='b')
ax2.tick_params(axis='y')
plt.show()
```

#### In the vehicles data set that you have downloaded, use the vehicles.csv file. In this exercise we will conduct some exploratory data analysis using one plot each of scatter plot, box plot, histogram, and bar chart. 

Scatter Plot

Please provide a scatter plot between "Combined MPG" as X variable and 
"Highway MPG" as Y variable

```python
# your code here

vehicles = pd.read_csv('vehicles.csv')

plt.scatter(vehicles['Combined MPG'], vehicles['Highway MPG'])
plt.title("Positive Correlation")
plt.xlabel("Combined MPG")
plt.ylabel("Highway MPG")
plt.show()
```

Box Whisker Plot

Please provide a box plot of the variable "CO2 Emission Grams/mile"

```python
# your code here



fig = sns.boxplot(vehicles['Year'], vehicles['CO2 Emission Grams/Mile'], data=vehicles)

fig.axis(ymin=0, ymax=vehicles['CO2 Emission Grams/Mile'].max())
fig.set_xticklabels(fig.get_xticklabels(),rotation=90)
```

Histogram

Please provide a histogram of the Fuel Barrels/Year

```python
# your code here

vehicles['Fuel Barrels/Year'].hist()
```

Bar Chart

Please provide a bar chart of the Fuel Type on the X axis and "City MPG" on the Y axis

```python
# your code here

ax = sns.barplot(x="Fuel Type", y="City MPG", data = vehicles)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
```
