### Bonus Question

#### For each department, compute the difference between the maximal salary and the minimal salary.

##### * Hint: try using `agg` or `apply` and `lambda`*

```python
# your answer h

#creation of difference function
def difference(x):
    return(max(x)-min(x))

#implementation of "difference" in groupby method
employee.groupby(['Department']).agg({'Salary':difference, 'Years':difference})
```

#### How many customers do we have in Spain?
##### Hint : Use value_counts()

```python
# your answer here

#value_counts returns a count of the unique values (like set) in CustomerID field, after calculating len(x) we have the solution.
x = orders[orders.Country == "Spain"]['CustomerID'].value_counts()
len(x)

#returns length of the set which is composed of only unique values, in this case the CustomerIDs for Spain.
len(set(orders[orders.Country == "Spain"]['CustomerID']))

#using .nunique() fx :

orders[orders.Country == "Spain"]['CustomerID'].nunique()
```

#### What's the  average amount of money spent by country

```python
# your answer here

#grouping by country and aggregating by their respective means'

avg_country = orders.groupby('Country', as_index=False).agg({'amount_spent':'mean'})

avg_country
```

#### What's the most expensive item?

```python
# your answer here

unit_price = orders.groupby('Description', as_index=False).agg({'UnitPrice':'max'})

expensive = unit_price.sort_values('UnitPrice', ascending = False)

expensive.head(1)
```

#### What was the average amount spent per year ?

```python
# your answer here

yearly_avg = orders.groupby('year', as_index=False).agg({'amount_spent':'mean'})

yearly_avg
```
