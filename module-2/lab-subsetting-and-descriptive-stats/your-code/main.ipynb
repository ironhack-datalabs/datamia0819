#### Find the mean and the standard deviation of the temperature of each state.

*Hint: Use functions from Data Manipulation lesson*

```python
# your answer here
avg_temp = temp.groupby('State', as_index=False).agg({'Temperature':'mean'})
avg_temp1 = temp.groupby('State', as_index=False).agg({'Temperature':'std'})

# print(avg_temp,"Mean", '\n\n', avg_temp1, "Stdev")
data3 = pd.merge(avg_temp, avg_temp1, on='State')
data3
```
