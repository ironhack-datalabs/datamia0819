```python
import pandas as pd
import numpy as np
```

```python
trx = pd.read_csv("creditcardcsvpresent.csv")
trx = trx.drop(columns = ["Transaction date","Merchant_id"])
```

```python
trx.isHighRiskCountry.value_counts()
```

```python
trx["isDeclined"] = trx["Is declined"]
trx["isFraudulent"] = trx["isFradulent"]
```

```python
trx = trx.drop(columns = ["Is declined", "isFradulent"])
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(trx.isForeignTransaction)

trx.isForeignTransaction = le.transform(trx.isForeignTransaction)
```

```python
le = preprocessing.LabelEncoder()
le.fit(trx.isFraudulent)

trx.isFraudulent = le.transform(trx.isFraudulent)
```

```python
le = preprocessing.LabelEncoder()
le.fit(trx.isDeclined)

trx.isDeclined = le.transform(trx.isDeclined)
```

```python
le = preprocessing.LabelEncoder()
le.fit(trx.isHighRiskCountry)

trx.isHighRiskCountry = le.transform(trx.isHighRiskCountry)
```
